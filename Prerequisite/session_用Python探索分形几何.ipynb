## ÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêà

ÂΩ¢Âºè‰∏äÔºåÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÊòØ‰∏Ä‰∏™Êó†ÈôêÁöÑÂ§çÊï∞ÈõÜÂêàÔºåËøô‰∫õÂ§çÊï∞Êª°Ë∂≥Âú®ÁªôÂÆöËø≠‰ª£Êù°‰ª∂‰∏ãÔºå‰ΩøÂ∫èÂàó `z0,z1, ...,zn, ...` ‰∏çÂèëÊï£ÁöÑÊÄßË¥®„ÄÇ

‰πüÂ∞±ÊòØËØ¥ÔºåË¶ÅÁ°ÆÂÆöÊüê‰∏™Â§çÊï∞`c`ÊòØÂê¶Â±û‰∫éÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêàÔºåÂè™ÈúÄÂ∞ÜËØ•Êï∞Â≠ó‰ª£ÂÖ•Âà∞‰∏äÈù¢ÁöÑÂÖ¨Âºè‰∏≠ÔºåÁÑ∂Âêé‰∏çÊñ≠Âú∞ËÆ°ÁÆó`zn`„ÄÇ

Â∫èÂàóÁöÑÁ¨¨‰∏Ä‰∏™ÂÖÉÁ¥†Ôºå`z0`ÔºåÂßãÁªàÁ≠â‰∫éÈõ∂„ÄÇËÆ°ÁÆó‰∏ã‰∏Ä‰∏™ÂÖÉÁ¥†Ôºå`z1 = c`ÔºåÁªßÁª≠Âπ≥Êñπ`z1`Âπ∂Âä†‰∏ä`c`ÔºåÂæóÂà∞`z2 = c¬≤ + c`Ôºå‰ª•Ê≠§Á±ªÊé®„ÄÇ

‰ΩÜÊòØÊàë‰ª¨‰∏ç‰ºöÔºà‰πüÊó†Ê≥ïÔºâÊó†ÈôêËø≠‰ª£‰∏ãÂéªÔºåÂΩì`n`Ë∂≥Â§üÂ§ßÊó∂ÔºåÊàë‰ª¨Â∞±ËÉΩËßÇÂØüÂà∞Â∫èÂàóÊòØÂê¶Êî∂Êïõ‰∫ÜÔºå‰πüÂç≥Â∫èÂàóÊòØÂê¶Ë∂ã‰∫é‰∏Ä‰∏™Á®≥ÂÆöÂÄº„ÄÇ

Â¶ÇÊûúÂ∫èÂàóÊ≤°ÊúâÂèëÊï£ÔºåÈÇ£‰πàÊàë‰ª¨Â∞±ÂèØ‰ª•ËØ¥`c`Â±û‰∫éÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêàÔºõÂèç‰πãÔºå`c`Â∞±‰∏çÂ±û‰∫éÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêà„ÄÇ

### ÂàùÊ≠•‰ª£Á†ÅÂÆûÁé∞

Êàë‰ª¨ÂÖàÊääËøô‰∏™ÈÄíÂΩíÂÖ¨ÂºèÊú¥Á¥†Âú∞ÂÜôÊàê `Python` ‰ª£Á†ÅÔºåÂπ∂Â∏¶ÂÖ•ÂÆûÊï∞ÁúãÁúãÔºö

````python
def iter_m(c):
    z = c
    for i in range(0, 10):  # Ëø≠‰ª£10Ê¨°
        z = z**2 + c
        print(round(z, 3), end=" -> ")
    print("\n" + "-" * 100)
````

````python
iter_m(-1)
iter_m(0.5)
iter_m(-0.5)
````

ÂèØ‰ª•ÁúãÂà∞ÔºåÂΩì`c`‰∏∫`-1`Êó∂ÔºåÂ∫èÂàóÂú®`0`Âíå`-1`‰πãÈó¥ÈúáËç°ÔºåËÄåÂΩì`c`‰∏∫`0.5`Êó∂ÔºåÂ∫èÂàóÂ¢ûÈïøÂø´ÂæóÊÉä‰∫∫ÔºåÂΩì`c`‰∏∫`-0.5`Êó∂ÔºåÂ∫èÂàóË∂ã‰∫é‰∏Ä‰∏™Á∫¶‰∏∫`-0.37`ÁöÑÁ®≥ÂÆöÂÄº„ÄÇ

Âõ†Ê≠§Êàë‰ª¨ÂèØ‰ª•ËØ¥`-0.5`ÂèØËÉΩÂ±û‰∫éÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêàÔºåËÄå`-1`Âíå`0.5`Âàô‰∏çÂ±û‰∫é„ÄÇ

Êé•‰∏ãÊù•Êàë‰ª¨ËØïÁùÄÂ∏¶ÂÖ•Â§çÊï∞„ÄÇ

> Ê≥®Ôºö Âú®Â§çÂπ≥Èù¢‰∏äÊï¥‰∏™ÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêàÂú®‰∏Ä‰∏™ÂçäÂæÑ‰∏∫ 2 ÁöÑÂúÜÂÜÖ„ÄÇ
>
> ËøôÊòØ‰∏Ä‰∏™Â∑≤ËØÅÊòéÁöÑÁªìËÆ∫ÔºåÂèØ‰ª•ÂÄüÊ≠§Ë∑≥ËøáËÆ∏Â§ö‰∏çÂøÖË¶ÅÁöÑËÆ°ÁÆóÔºå`|z| > 2` ÁöÑÁÇπËÇØÂÆö‰∏çÂ±û‰∫éËØ•ÈõÜÂêà„ÄÇ

‰øÆÊîπ‰ª£Á†ÅÂ¶Ç‰∏ãÔºö

````python
def iter_m(c):
    z = c
    for i in range(0, 10):  # Ëø≠‰ª£10Ê¨°
        if(abs(z) > 2):
            break # ÂçäÂæÑÂ§ß‰∫é2ÂàôËÆ§‰∏∫ÂèëÊï£
        z = z**2 + c
        print(z)

iter_m(0.1+0.2j)
````

Êàë‰ª¨ÁúãÂà∞ÔºåÂ∫èÂàóÊî∂ÊïõÁöÑË∂ãÂäøÂçÅÂàÜÊòéÊòæÔºåÂõ†Ê≠§Êàë‰ª¨ÂèØ‰ª•ËØ¥Ôºå`0.1+0.2j` ÂæàÂ§ßÂèØËÉΩÂ±û‰∫éÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêà„ÄÇ

### ‰ªéÈõÜÂêàÂà∞ÂõæÂÉè

Êàë‰ª¨ÈÉΩÁü•ÈÅìÔºå‰∏Ä‰∏™Â§çÊï∞`a+bj`‰∏éÂ§çÂπ≥Èù¢‰∏äÁöÑÁÇπ`(a, b)`ÂîØ‰∏ÄÂØπÂ∫î„ÄÇ

Êé•‰∏ãÊù•Êàë‰ª¨Ë¶ÅÂÅöÁöÑÂ∞±ÊòØÂú®Â§çÂπ≥Èù¢‰∏äÔºåÂ∞ÜÂ±û‰∫éÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêàÁöÑÁÇπÂíå‰∏çÂ±û‰∫éËØ•ÈõÜÂêàÁöÑÁÇπ‚ÄúÊüì‰∏ä‚Äù‰∏çÂêåÁöÑÈ¢úËâ≤ÔºåÂ∞±ËÉΩÁúãÂà∞Áõ¥ËßÇÁöÑÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêàÁöÑÂõæÂÉè‰∫Ü„ÄÇ

ËôΩÁÑ∂ÁõÆÂâçÂÆÉÂè™Êúâ‰∏§ÁßçÈ¢úËâ≤Ôºå‰ΩÜËøô‰πüÊòØÊàë‰ª¨ÂêëÂâçÊé¢Á¥¢ÁöÑÈáçË¶Å‰∏ÄÊ≠•„ÄÇ

Êàë‰ª¨ÈúÄË¶ÅÂÖà‰øÆÊîπ`iter_m(c)`ÂáΩÊï∞ÔºåËÆ©ÂÆÉËøîÂõû‰∏Ä‰∏™`bool`ÂÄº:

````python
def iter_m(c):
    z = c
    for i in range(0, 200):  # Ëø≠‰ª£200Ê¨°
        if(abs(z) > 2):
            return False # ÂçäÂæÑÂ§ß‰∫é2ÂàôËÆ§‰∏∫ÂèëÊï£
        z = z**2 + c
    return True

iter_m(0.1+0.2j)
````

Ê≥®ÊÑèÂà∞ÊàëËøò‰øÆÊîπ‰∫ÜËø≠‰ª£Ê¨°Êï∞Ôºå‰ª•‰æøÂ∞ΩÂèØËÉΩÂæóÂà∞ÂáÜÁ°ÆÁöÑÁªìÊûú„ÄÇ

Â¶ÇÊûúËø≠‰ª£`200`Ê¨°Âêé‰ªçÁÑ∂Ê≤°ÊúâÂèëÊï£ÔºåÈÇ£‰πàÊàë‰ª¨Â∞±ËÆ§‰∏∫Â§çÊï∞`c`Â±û‰∫éÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêà„ÄÇ

Ê≠£Â¶ÇÊàë‰ª¨ÂâçÈù¢ÊâÄÁåúÊµãÁöÑÔºå`0.1+0.2j`ÁöÑÁ°ÆÂ±û‰∫éÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêà„ÄÇ

‰Ω†ÂèØ‰ª•Ëá™Â∑±Â∞ùËØïÂ∏¶ÂÖ•Êõ¥Â§öÁöÑÂ§çÊï∞ÔºåÁúãÁúãÂÆÉ‰ª¨ÊòØÂê¶Â±û‰∫éÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêà„ÄÇ

Ë¶ÅÁªòÂà∂ÂõæÂÉèÔºåÊàë‰ª¨ËøòÈúÄË¶Å‰∏ÄÂº†‚ÄúÁôΩÁ∫∏‚ÄùÔºåÊàëÂáÜÂ§áÂú®Ëá™ÂÆö‰πâÁöÑÁªòÂõæÂáΩÊï∞‰∏≠ÂÆûÁé∞ÂÆÉÔºö

````python
import numpy as np
import pylab as pl
from matplotlib import cm


def draw_set(cx, cy, d, ufunc: np.ufunc):
    x0, x1, y0, y1 = cx - d, cx + d, cy - d, cy + d
    y, x = np.ogrid[y0:y1:400j, x0:x1:400j] # Ëøô‰∏ÄË°åÂ∞ÜÊï¥Âº†ÁôΩÁ∫∏ÂàÜ‰∏∫‰∫Ü`400*400`ÁöÑÂ∞èÊ†º(ÂÉèÁ¥†ÁÇπ)Ôºå`400j`Ë°®Á§∫Êää`y0`Âà∞`y1`(Êàñ`x0`Âà∞`x1`)Âπ≥ÂùáÂàÜ‰∏∫400‰ªΩ„ÄÇ
    z = x + y * 1j
    pl.imshow(ufunc(z).astype(float), cmap=cm.jet, extent=[x0, x1, y0, y1])
````

#### ÂáÜÂ§á‚ÄúÁîªÁ∫∏‚Äù

`(cx, cy)`ÊòØËøôÂº†‚ÄúÁôΩÁ∫∏‚ÄùÁöÑ‰∏≠ÂøÉÔºå`d`ÊòØ‰ªé‰∏≠ÂøÉÂêë‰∏§ËæπÂª∂‰º∏ÁöÑÈïøÂ∫¶Ôºå‰πüÂ∞±ÊòØËØ¥ÔºåËøôÂº†‚ÄúÁôΩÁ∫∏‚ÄùÊòØ‰∏Ä‰∏™‰ª•`(cx, cy)`‰∏∫‰∏≠ÂøÉÔºåËæπÈïø‰∏∫`2d`ÁöÑÊ≠£ÊñπÂΩ¢Âå∫Âüü„ÄÇ

`y, x = np.ogrid[y0:y1:400j, x0:x1:400j]` Ëøô‰∏ÄË°åÂ∞ÜÊï¥Âº†ÁôΩÁ∫∏ÂàÜ‰∏∫‰∫Ü`400*400`ÁöÑÂ∞èÊ†º(ÂÉèÁ¥†ÁÇπ)Ôºå`400j`Ë°®Á§∫Êää`y0`Âà∞`y1`(Êàñ`x0`Âà∞`x1`)Âπ≥ÂùáÂàÜ‰∏∫400‰ªΩ„ÄÇ

ÂèØ‰ª•‰ªé‰∏ãÈù¢ÁöÑÁ§∫‰æã‰ª£Á†Å‰∏≠ÁúãÂá∫`np.ogrid`ÊòØÂ¶Ç‰ΩïËµ∑‰ΩúÁî®ÁöÑ:

````python
y,x = np.ogrid[-1:1:5j, -1:1:5j]
print(x)
print(y)
````

Êé•‰∏ãÊù•ÊûÑÈÄ†‰∫ÜÂ§çÊï∞`z = x + y*1j`ÔºåÂ∞ÜËøôÂº†ÁôΩÁ∫∏‰∏éÊàë‰ª¨ÁöÑÂ§çÂπ≥Èù¢ËÅîÁ≥ªËµ∑Êù•ÔºåÂè™Áî®‰∏Ä‰∏™Â§çÊï∞Êï∞ÁªÑÔºåÂ∞±ËÉΩË°®Á§∫Ëøô‰∫õ`x*y`„ÄÇ

Âú®ËøôÈáåÔºåÂõ†‰∏∫`x`ÁöÑÂΩ¢Áä∂ÊòØ`1*5`Ôºå`y`ÁöÑÂΩ¢Áä∂ÊòØ`5*1`ÔºåÂõ†Ê≠§`z`ÁöÑÂΩ¢Áä∂ÊòØ`5*5`Ôºö

````python
z = x + y * 1j
print(z) # 5*5ÁöÑÁü©Èòµ
````

Êàë‰ª¨Âú®ÂÆûÈôÖÁªòÂõæÁöÑÊó∂ÂÄôÔºå‰∏∫‰∫ÜËÆ©ÂõæÁâáÊõ¥Ê∏ÖÊô∞Ôºå`z`ÁöÑÂ§ßÂ∞èÊòØ`400*400`„ÄÇ

#### ËÆ©Ëø≠‰ª£ÂáΩÊï∞Áõ¥Êé•ÊîØÊåÅÊï∞ÁªÑËøêÁÆó

`pl.imshow(ufunc(z).astype(float), cmap=cm.jet, extent=[x0, x1, y0, y1])` Ëøô‰∏ÄË°åÊòØÁªòÂõæÁöÑ‰∏ªË¶Å‰ª£Á†Å„ÄÇ

`ufunc`ÊòØ‰∏Ä‰∏™`np.ufunc`Á±ªÂûãÁöÑÂáΩÊï∞ÔºåËÉΩÂ§üÊñπ‰æøÂú∞Â§ÑÁêÜÊï∞ÁªÑÁ±ªÂûãÁöÑÂèÇÊï∞„ÄÇ`cmap`ÊòØÈÖçËâ≤ÊñπÊ°àÔºåÊàëÈÄâÊã©ÁöÑÊòØ`jet`„ÄÇ`extent`ÊòØÁªòÂõæÂå∫Âüü„ÄÇ

ÂÖ∑‰ΩìÊù•ËØ¥ÔºåÊàë‰ª¨ÂâçÈù¢ÂÆö‰πâÁöÑ`iter_m(c)`ÂáΩÊï∞ÔºåÂè™ËÉΩÂ§ÑÁêÜÂçï‰∏™Êï∞Â≠óÔºåÊó†Ê≥ïÁõ¥Êé•Â§ÑÁêÜÊï∞ÁªÑÔºåËÄåÁé∞Âú®ÁöÑ`z`Â∞±ÊòØ‰∏Ä‰∏™Â§çÊï∞Êï∞ÁªÑ„ÄÇ

‰∏∫‰∫ÜÂ∞Ü‰∏Ä‰∏™ÊôÆÈÄöÁöÑ`Python`ÂáΩÊï∞Âèò‰∏∫‰∏Ä‰∏™ÂèØ‰ª•Áõ¥Êé•Â§ÑÁêÜÊï∞ÁªÑÁöÑÂáΩÊï∞ÔºåÊàë‰ª¨ÂèØ‰ª•‰ΩøÁî®`numpy`Êèê‰æõÁöÑ`frompyfunc(func, nin, nout) -> ufunc`ÊñπÊ≥ï„ÄÇ

ÂÖ∂‰∏≠funcÊòØÊàë‰ª¨Ëá™ÂÆö‰πâÁöÑÂáΩÊï∞Ôºå`nin`ÊòØ‰º†ÂÖ•ÂèÇÊï∞ÁöÑ‰∏™Êï∞Ôºå`nout`ÊòØ‰º†Âá∫ÁªìÊûúÁöÑ‰∏™Êï∞„ÄÇ

ÂÖ∑‰Ωì‰ΩøÁî®ÊñπÊ≥ïÂ¶Ç‰∏ãÔºö

````python
mandelbrot = np.frompyfunc(iter_m, 1, 1)
mandelbrot(z)
````

ÂèØ‰ª•ÁúãÂà∞ÔºåÊàë‰ª¨ÊàêÂäüÂú∞Â∞Ü`z`ÁöÑÊØè‰∏Ä‰∏™ÂùêÊ†áÈÉΩÂà§Êñ≠Âá∫‰∫ÜÊòØÂê¶Â±û‰∫éÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêà„ÄÇ

Êàë‰ª¨ÁîªÂá∫Ëøô‰∏™Âè™Êúâ5*5ÂÉèÁ¥†ÁöÑÂõæÁâáÁúãÁúãÔºö

ÂóØÔºåÊûúÁÑ∂ÈùûÂ∏∏ÁöÑÊäΩË±°ÔºåÁúãÊù•`400*400`ÁöÑÂàÜËæ®ÁéáÊòØÂæàÊúâÂøÖË¶ÅÁöÑ„ÄÇ

Áé∞Âú®Ôºå‰∏á‰∫ãÂÖ∑Â§áÔºåÂè™Ê¨†ËæìÂá∫„ÄÇÂáÜÂ§áÂ•ΩÁîªÂá∫Êàë‰ª¨ÁöÑÁ¨¨‰∏ÄÂº†ÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêàÁöÑÂõæÂÉè‰∫Ü„ÄÇ

### ‰ªéÂèåËâ≤Âà∞ÂΩ©Ëâ≤

Êàë‰ª¨ÊàêÂäü‰∫ÜÔºå‰ΩÜÊòØÂè™Êúâ‰∏§ÁßçÈ¢úËâ≤Ôºå‰∏çÂ§üÈúáÊíºÔºåÊÉ≥Ë¶ÅÁîªÂá∫ÂΩ©Ëâ≤ÁöÑÂõæÂÉèÊÄé‰πàÂäûÂë¢Ôºü

Êàë‰ª¨ÈúÄË¶ÅÂÜçÊ¨°‰øÆÊîπ`iter_m(c)`ÔºåËÆ©ÂÆÉ‰∏çÂè™ËøîÂõû`True/False`‰∏§‰∏™ÂÄº„ÄÇ

ÂÖàÁúãÁúãÁé∞Âú®ÁöÑÁâàÊú¨Ôºö

```python
def iter_m(c):
    z = c
    for i in range(0, 200):  # ÊúÄÂ§öËø≠‰ª£200Ê¨°
        if abs(z) > 2:
            return False  # ÂçäÂæÑÂ§ß‰∫é2ÂàôËÆ§‰∏∫ÂèëÊï£
        z = z * z + c
    return True
```

ÊÄé‰πàÊîπÂë¢Ôºü

‰∏Ä‰∏™ÂæàÂ•ΩÁöÑideaÊòØÁõ¥Êé•ËøîÂõûËø≠‰ª£Ê¨°Êï∞`i`Ôºå‰øÆÊîπÂêéÁöÑ`iter_m(c)`Â¶Ç‰∏ãÔºö

````python
def iter_m(c):
    z = c
    for i in range(0, 200):  # Ëø≠‰ª£200Ê¨°
        if(abs(z) > 2):
            break # ÂçäÂæÑÂ§ß‰∫é2ÂàôËÆ§‰∏∫ÂèëÊï£
        z = z**2 + c
    return i
````

ÊÉ≥ÊÉ≥ÁúãÔºåËøîÂõû`i`Â¶ôÂú®Âì™ÈáåÔºü

Êé•‰∏ãÊù•Â∞±ÊòØÈáçÂ§çÂâçÈù¢ÁöÑÁªòÂõæÊ≠•È™§ÔºåÊàë‰ª¨Â∞±ËÉΩÂæóÂà∞‰∏ÄÂº†ÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêàÁöÑÂΩ©ÁÖß‰∫ÜÔºö

````python
mandelbrot = np.frompyfunc(iter_m, 1, 1)
draw_set(-0.5, 0, 1.0, mandelbrot)
````

Êàë‰ª¨ËøòÂèØ‰ª•ÈÄöËøá‰∏ÄÊÆµÈÄêÁ∫ßÊîæÂ§ßÁöÑ‰ª£Á†ÅÊù•ÁúãÁúãÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÁöÑÁªÜËäÇÔºö

````python
x, y = 0.2732, 0.5951
pl.subplot(231)
draw_set(-0.5, 0, 1, mandelbrot)
for i in range(2, 7):
    pl.subplot(230 + i)
    draw_set(x, y, 0.25**(i-1), mandelbrot)

pl.subplots_adjust(0.05, 0.1, 0.95, 1, 0.5, 0.1)
````

‰ªñÂú®Ê≤°ÊúâËÆ°ÁÆóÊú∫ÁöÑÂ∏ÆÂä©‰∏ãÂèëÁé∞‰∫ÜJuliaÈõÜÂêà„ÄÇ

ÂõûÈ°æ‰∏Ä‰∏ãÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêàÁöÑÈÄíÂΩíÊñπÂºèÔºö

ËÄåË¶ÅÁ°ÆÂÆö‰∏Ä‰∏™Â§çÊï∞`z`ÊòØÂê¶Â±û‰∫éJuliaÈõÜÂêàÔºåÂøÖÈ°ª‰ΩøÁî®ËØ•Êï∞Â≠ó‰Ωú‰∏∫Â∫èÂàóÁöÑËµ∑ÁÇπÔºåÂπ∂‰∏∫`c`ÂèÇÊï∞ÈÄâÊã©Âè¶‰∏Ä‰∏™Á°ÆÂÆöÁöÑÂÄº„ÄÇ

ÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêàÂè™Êúâ‰∏Ä‰∏™ÔºåËÄåJuliaÈõÜÂêàÊúâÊó†ÈôêÂ§ö‰∏™„ÄÇ

Êàë‰ª¨Âè™ÈúÄË¶ÅÂØπÊõºÂæ∑Â∏ÉÊ¥õÁâπÈõÜÂêàÁöÑËø≠‰ª£ÂáΩÊï∞ËøõË°åÁÆÄÂçïÁöÑ‰øÆÊîπÔºåÂ∞±ÂèØ‰ª•Â∫îÁî®Âú®JuliaÈõÜÂêà‰∏äÔºö

````python
def iter_j(z):
    c = -0.4 + 0.6j
    for i in range(0, 200):  # Ëø≠‰ª£200Ê¨°
        if(abs(z) > 2):
            break # ÂçäÂæÑÂ§ß‰∫é2ÂàôËÆ§‰∏∫ÂèëÊï£
        z = z**2 + c
    return i
````

Êé•‰∏ãÊù•Êàë‰ª¨Áî®ÂíåÂâçÈù¢Áõ∏ÂêåÁöÑ‰∏ÄÂ•óÁªòÂõæÊñπÊ≥ïÔºåÂ∞±ËÉΩÂæóÂà∞JuliaÈõÜÂêàÁöÑÂΩ©ÁÖß‰∫ÜÔºö

````python
julia = np.frompyfunc(iter_j, 1, 1)
draw_set(0, 0, 1.5, julia)
````

ÂêåÊ†∑ÂèØ‰ª•ÈÄöËøá‰∏ÄÊÆµÈÄêÁ∫ßÊîæÂ§ßÁöÑ‰ª£Á†ÅÊù•ÁúãÁúãJuliaÈõÜÁöÑÁªÜËäÇÔºö

````python
x, y = 0.5754, 0.2048
pl.subplot(231)
draw_set(0, 0, 1.5, julia)
for i in range(2, 7):
    pl.subplot(230 + i)
    draw_set(x, y, 0.25 ** (i - 1), julia)
pl.subplots_adjust(0.05, 0.1, 0.95, 1, 0.5, 0.1)
````
