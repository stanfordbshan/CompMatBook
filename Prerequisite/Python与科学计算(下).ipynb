### å››ã€åŸºæœ¬æ“ä½œ
æ•°ç»„çš„ç®—æ•°è®¡ç®—æ˜¯åœ¨å…ƒç´ å±‚çº§è¿ç®—çš„ã€‚è®¡ç®—ç»“æœä¼šå­˜åœ¨ä¸€ä¸ªæ–°åˆ›å»ºçš„æ•°ç»„ä¸­ã€‚

åœ¨NumPyä¸­```*```å·ä»ç„¶è¡¨ç¤ºä¹˜æ³•ï¼ŒçŸ©é˜µä¹˜ç§¯ç”¨```np.dot```æ¥è®¡ç®—ã€‚

````python
A=np.array([(1,1),(0,1)])
B=np.array([(2,0),(3,4)])
print(A*B)  #å¯¹åº”æ•°å­—ç›¸ä¹˜
print(A.dot(B))
print(np.dot(A,B))   #çŸ©é˜µç›¸ä¹˜
````

ç±»ä¼¼äº```+=```å’Œ```*=```çš„è¿ç®—æ˜¯ç›´æ¥åœ¨ç°æœ‰æ•°ç»„ä¸Šè®¡ç®—çš„ï¼Œæ²¡æœ‰åˆ›å»ºæ–°çš„æ•°ç»„ã€‚Numpyä¸­çš„è®¡ç®—åŒæ ·ä¹Ÿæ˜¯å‘ä¸Šè½¬å‹çš„ï¼Œå¯ä»¥ç®€å•ç†è§£æˆæµ®ç‚¹æ•°å’Œæ•´æ•°è¿ç®—çš„ç»“æœæ˜¯æµ®ç‚¹æ•°ã€‚

````python
a = np.ones((2,3), dtype=int)
b = np.random.random((2,3))
a*=3
print(a)
b += a
print(b)
# a += b                  # æµ®ç‚¹æ•°ä¸ä¼šè‡ªåŠ¨è½¬æ¢æˆæ•´æ•°
````

```np.ndarray```æä¾›äº†è®¸å¤šä¸€å…ƒæ“ä½œã€‚æ¯”å¦‚æ•°ç»„æ±‚å’Œã€æ±‚æœ€å¤§æœ€å°å€¼ç­‰ã€‚

````python
a=np.random.random((2,3))
print(a)
print(a.sum())
print(a.mean())
print(a.max())
print(a.min())
````

é»˜è®¤çš„ï¼Œè¿™äº›ä¸€å…ƒæ“ä½œæ˜¯å¯¹æ•´ä¸ªæ•°ç»„è¿›è¡Œè®¡ç®—ï¼Œæ²¡æœ‰è€ƒè™‘åˆ°æ•°ç»„çš„å½¢çŠ¶ã€‚ä½ å¯ä»¥è®¾ç½®```axis```å‚æ•°æ¥æŒ‡å®šè¿ç®—æ–¹å‘ã€‚```axis```è¡¨ç¤ºç¬¬nç»´ï¼ˆä»0å¼€å§‹ï¼‰ã€‚

````python
b=np.arange(12).reshape(3,4)
print(b)
print(b.sum(axis=0)) #å¯¹ç¬¬0ç»´çš„å…ƒç´ æ±‚å’Œ
print(b.sum(axis=1)) #å¯¹ç¬¬1ç»´çš„å…ƒç´ æ±‚å’Œ
print(b.min(axis=1))
print(b.cumsum(axis=1)) #å¯¹ç¬¬1ç»´çš„å…ƒç´ ä¾æ¬¡ç´¯åŠ æ±‚å’Œ
````

### äº”ã€å¹¿æ’­å‡½æ•°
NumPyæä¾›äº†ç†ŸçŸ¥çš„æ•°å­¦æ–¹æ³•ï¼Œå¦‚ï¼šsinã€cosã€expç­‰ã€‚åœ¨NumPyä¸­ï¼Œè¿™äº›æ–¹æ³•è¢«ç§°ä½œå¹¿æ’­å‡½æ•°ã€‚è¿™äº›å‡½æ•°ä¼šå¯¹æ•°ç»„ä¸­çš„æ¯ä¸ªå…ƒç´ è¿›è¡Œè®¡ç®—ï¼Œè¿”å›è®¡ç®—åçš„æ•°ç»„ã€‚

````python
B=np.arange(3)
print(B)
print(np.exp(B))
print(np.sqrt(B))
C=np.array([2,-1,4])
print(np.add(B,C))
print(B+C)
````

### å…­ã€ç´¢å¼•ã€åˆ‡ç‰‡å’Œè¿­ä»£
ä¸€ç»´æ•°ç»„å¯ä»¥è¢«ç´¢å¼•ã€åˆ‡ç‰‡å’Œè¿­ä»£ï¼Œå°±å’ŒPythonä¸­çš„åˆ—è¡¨ä¸€æ ·ã€‚

å¤šç»´æ•°ç»„å¯ä»¥åœ¨æ¯ä¸€ä¸ªç»´åº¦æœ‰ä¸€ä¸ªç´¢å¼•ï¼Œè¿™äº›ç´¢å¼•æ„æˆå…ƒç»„æ¥è¿›è¡Œè®¿é—®ã€‚
