### break

### continue
ä½¿å¾ªç¯è·³è¿‡å…¶ä¸»ä½“çš„å‰©ä½™éƒ¨åˆ†ï¼Œå¹¶ç«‹å³è¿›å…¥ä¸‹ä¸€æ¬¡å¾ªç¯ã€‚

### pass
å½“è¯­æ³•éœ€è¦ä½†ä¸éœ€è¦æ‰§è¡Œä»»ä½•å‘½ä»¤æˆ–ä»£ç æ—¶ï¼ŒPythonä¸­å°±å¯ä»¥ä½¿ç”¨passè¯­å¥ï¼Œæ­¤è¯­å¥ä»€ä¹ˆä¹Ÿä¸åšï¼Œç”¨äºè¡¨ç¤ºâ€œå ä½â€çš„ä»£ç ã€‚


````python
numbers = [1, 2, 3, 4, 5]
for num in numbers:
    if num % 2 == 0:
        continue
    print(num)
print("-" * 9)
for num in numbers:
    if num == 3:
        break
    print(num)
print("-" * 9)
for num in numbers:
    if num == 3:
        pass
    print(num)
print("-" * 9)
````
